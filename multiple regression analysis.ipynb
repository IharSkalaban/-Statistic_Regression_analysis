### Сначала построим небольшую модель всего с одним предиктором цены (price) – horsepower.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
results = smf.ols('price ~ horsepower', cars_lr).fit()
print(results.summary())
```

Данная модель объясняет 65% изменчивости цен.

```python
# Сделать интерплот понятнее
cars_lr['horsepower_cent'] = cars_lr['horsepower'] - cars_lr['horsepower'].mean() # центрирование

results_cent = smf.ols('price ~ horsepower_cent', cars_lr).fit()
print(results_cent.summary())
```

```python
1.328e+04 # price Это означает, что для автомобилей со средним horsepower по выборке цена будет составлять 13280$
```

### Теперь построим модель со всеми предикторами

```python
X = cars_lr.drop(['price','horsepower_cent'], axis='columns') # предикторы
y = cars_lr['price'] # целевая переменная
```

```python
# не забываем про константу
X = sm.add_constant(X)
```

```python
X
```

```python
model_1 = sm.OLS(y, X).fit()
print(model_1.summary())
```

#### Данная модель объясняет 95% изменчивости цен. Однако только 3 марки машин значимы в данной модели (p<0.05)

### Построим модель со всеми предикторами, кроме марок машин

```python
cols_to_keep = cars_lr.columns[~cars_lr.columns.str.startswith('company_')]
cols_to_keep
```

```python
# модель без марок
X_2 = cars_lr[cols_to_keep].drop('price', axis='columns')
y_2 = cars_lr['price']
```

```python
X_2 = sm.add_constant(X_2)
```

```python
model_2 = sm.OLS(y_2, X_2).fit()
print(model_2.summary())
```

#### Данная модель объясняет 90% изменчивости цен. Среди предикторов 10 из 27 оказались не значимыми (p > 0.05)


### Итог: так как Adj. R-squared в последних двух моделях изменился не значительно, то лучше выбрать модель без марок машин
